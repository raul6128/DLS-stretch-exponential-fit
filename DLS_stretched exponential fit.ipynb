```python
import pandas as pd
file_path = 'C:\\Users\\zl29b\\OneDrive - University of Glasgow\\Documents\\Wolfram Mathematica\\'
file_name = 'yang wang_dmem-5d'
# Read the Excel file into a DataFrame
df = pd.read_excel(file_path + file_name + '.xlsx',sheet_name='Measurement 0')
```

```python
# Convert `Unnamed: 13` (Delay time) and `Unnamed: 14` (g2) to numeric, setting failed conversions to NaN
df['delay_time'] = pd.to_numeric(df['Unnamed: 13'], errors='coerce')
df['g2'] = pd.to_numeric(df['Unnamed: 14'], errors='coerce')

# Filter the data between rows with index 9 to 211 (inclusive)
df_filtered = df.loc[9:211, ['Unnamed: 13', 'Unnamed: 14']]

# Display the first 5 rows of the extracted columns
print("Data from Delay time and g2:\n", df_filtered.head(5).to_markdown(index=False, numalign="left", stralign="left"))
```

```python
import matplotlib.pyplot as plt

# Extract the columns for plotting
x = df_filtered['Unnamed: 13']
y = df_filtered['Unnamed: 14']

# Create the plot
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
plt.scatter(x, y)
plt.xscale('log')  # Set x-axis to logarithmic scale
plt.xlabel('Delay time (s)')
plt.ylabel('g2')
plt.title('Scatter Plot of Delay time vs g2')
plt.grid(True)
plt.show()
```
